```python
import requests,json
import pandas as pd
```

```python
url = "http://127.0.0.1:8000/marketdept/"
Response = requests.get(url).content
data = json.loads(Response)
data = data["body"]["Data"]
print(data)
```

```python
df2 = pd.DataFrame.from_dict(data)
print(df2)
```

```python
df2.to_csv('file_name.csv', encoding='utf-8')
```

```python
df2.to_excel('Courses.xlsx')
```

```python
scripMaster = pd.read_csv('scripmaster-csv-format.csv')
```
